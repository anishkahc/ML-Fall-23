# Goyal Macro File

```python
#Read file and display info and 
goyal = pd.read_parquet(data_dir + goyal_file)
```

```python
goyal.head(5)
```

```python
#check for duplicates
goyal.duplicated().max()
```

```python
#1. Find no. of unique values in 'Date' field
goyal['Date'].nunique
```

```python
# define function to check if column is a valid primary key
def key_field_check(file, column_name):
    if file[column_name].duplicated().max()==True:print(True)
    else: print(False)
```

```python
key_field_check(file=goyal, column_name=['Date'])
```

```python
#3. Check for suggested index column 'Date'
def index_fn(file_name, column_names):
    length= len(column_names)
    if length == 1:
        file_name.set_index(keys=[column_names[0]], inplace=True, verify_integrity=True)
        file_name.sort_index(inplace=True)
        
    else:
        file_name.set_index(keys=[column_names[0], column_names[1]], inplace=True, verify_integrity=True)
        file_name.sort_index(inplace=True)
```

```python
index_fn(file_name = goyal, column_names = ['Date'])
print(goyal.index.name)
```

```python
#4. Create new column dp and store dividend-price ratio
goyal['dp'] = goyal['D12'] / goyal['Index']
```

```python
#5. Create new column ep with earnings-price ratio
goyal['ep'] = goyal['E12'] / goyal['Index']
```

```python
#6. Create new column tms to store term spread
goyal['tms'] = goyal['lty'] - goyal['tbl']
```

```python
#7. Create new column dfy to store default term spread
goyal['dfy'] = goyal['BAA'] - goyal['AAA']
```

```python
goyal.info(verbose= True, show_counts= True)
goyal.head(10)
```

```python
#8. Calculate correlation between S&P 500 Index and Corporate Bond Yields on -AAA rated bonds
corr= goyal['Index'].corr(goyal['AAA'])
print('Correlation between SP500 Index & Corp Bond yeild is ', corr)
```

```python
#9. Calculate volatility(standard deviation) of Treasury Bill yield
std_tbill = goyal['tbl'].std()
print(std_tbill)
```

```python
#10. Calculate volatility of t bill as a percentage of fraction of average treasury bill rate
avg_treasury_rate = goyal['Rfree'].mean()
volatility_fraction = std_tbill/avg_treasury_rate

print('Average treasury rate = ', avg_treasury_rate)
print('Volatility fraction of a average treasury bill rate = ', volatility_fraction)
```

```python
#3. Get sample stats like mean, median, std, min, max, quartiles for a. Monthly excess return, b. risk free rate

print('Monthly excess market return statistics', '\n')
print('Mean: ', ff3['Mkt-RF'].mean())
print('Median: ', ff3['Mkt-RF'].median())
print('Standard Deviation: ', ff3['Mkt-RF'].std())
print('Min: ', ff3['Mkt-RF'].min())
print('Max: ', ff3['Mkt-RF'].max())
print('Quartiles: ', ff3['Mkt-RF'].quantile([0.25, 0.5, 0.75]))

```

```python
print('Risk Free Rate statistics', '\n')
print('Mean: ', ff3['RF'].mean(),)
print('Median: ', ff3['RF'].median())
print('Standard Deviation: ', ff3['RF'].std())
print('Min: ', ff3['RF'].min())
print('Max: ', ff3['RF'].max())
print('Quartiles: ', ff3['RF'].quantile([0.25, 0.5, 0.75]))
```

```python
# What do they suggest for the predictability of market portfolio and risk-free rate?

print('The mean, median and std are higher for the market portfolio than the risk free rate. The mean and std are used to estimate the range of changes in the market portfolio and risk free rate but the min and max values establish the range.')
print('Minimum value of market portfolio < 0 which mean that the reutrn was less than the risk free rate. This indicates negative returns.')
print('Quartiles help divide the values such that we can predict values in a particular quartile')
```
