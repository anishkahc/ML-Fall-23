```python
#3. Check for suggested index column 'Date'
def index_fn(file_name, column_names):
    length= len(column_names)
    if length == 1:
        file_name.set_index(keys=[column_names[0]], inplace=True, verify_integrity=True)
        file_name.sort_index(inplace=True)
        
    else:
        file_name.set_index(keys=[column_names[0], column_names[1]], inplace=True, verify_integrity=True)
        file_name.sort_index(inplace=True)
```

```python
goyal['dp'] = goyal['D12'] / goyal['Index']
goyal['ep'] = goyal['E12'] / goyal['Index']
goyal['tms'] = goyal['lty'] - goyal['tbl']
goyal['dfy'] = goyal['BAA'] - goyal['AAA']
```

```python
goyal = goyal.rename(columns = {'Date':'month'})
goyal['month'] = pd.to_datetime(goyal['month'], format = "%Y%m")
goyal['month'] = goyal['month'] + pd.tseries.offsets.MonthEnd(0)

#1.1 creating variable month_lagged
goyal['month_lagged'] = goyal['month'] + pd.tseries.offsets.MonthEnd(-1)
goyal.head()
```

```python
#1.2 merge msf with goyal
crsp_msf = pd.read_parquet(data_dir + crsp_msf_file)
crsp_msf['mvel']= crsp_msf['prc'] * crsp_msf['shrout']

index_fn(file_name = crsp_msf, column_names = ['permno','date'])

df_merged1 = pd.merge(left = crsp_msf.reset_index(), right = goyal, how = 'left',
                     left_on = ['date'], right_on = ['month_lagged'],
                      indicator = True, validate = 'm:1')

df_merged1._merge.value_counts()
df_merged1.drop(axis = 1, columns = '_merge', inplace = True)
```

```python
#2 left join df_merged1 with ff3
ff3 = pd.read_parquet(data_dir + ff3_file)
```

```python
df_merged2 = pd.merge(left = df_merged1.reset_index(), right = ff3.reset_index(), 
                      how = 'left', left_on = ['date'], right_on = ['Date'],
                     indicator = True, validate = 'm:1')
df_merged2.drop(axis = 1, columns = '_merge', inplace = True)
df_merged2.head()
```

```python
#3.1
crsp_delisted = pd.read_parquet(data_dir + crsp_delist_file)

delisted = crsp_delisted['dlret'].isna() & crsp_delisted['dlstcd'].between(400, 591)
crsp_delisted.loc[delisted, 'dlret']= -0.3
# print(crsp_delisted)
```

```python
df_merged3 = pd.merge(left = df_merged2.reset_index(), right = crsp_delisted.reset_index(), 
                      how = 'left', 
                      left_on = ['permno','date'], right_on = ['permno', 'dlstdt'],
                     indicator = True, validate = 'm:1')
df_merged3.drop(axis = 1, columns = '_merge', inplace = True)
df_merged3.head()
```

```python
#3.3
df_merged3['final_return'] = (1 + df_merged3['ret'])*((1 + df_merged3['dlret'])) - 1
#df_merged3.head()
```

```python
#4.1
open_ap = pd.read_parquet(data_dir + open_ap_file)
open_ap['Date'] = open_ap['Date']  + pd.tseries.offsets.MonthEnd(0)
open_ap['month_lagged'] = open_ap['Date']  + pd.tseries.offsets.MonthEnd(-1)
index_fn(file_name = open_ap, column_names=['permno','Date'])
```

```python
#4.2
df_merged_all = pd.merge(left=df_merged3, 
                     right=open_ap, 
                     how='inner', 
                     left_on=['permno', 'month_lagged'], right_on=['permno', 'month_lagged'],
                     indicator = True, validate = 'm:1')
```

```python
#4.3
df_merged_all['excess_return'] = df_merged_all['ret'] - df_merged_all['RF']
# df_merged_all.head()
```

```python
#5.1 Sample Statistics
not_na_df = df_merged_all[df_merged_all['excess_return'].notna()]
not_na_df.groupby('month')['permno'].nunique()
```
